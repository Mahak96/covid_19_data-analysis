```python

covid.drop(['Sno', 'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace=True, axis=1)

```

```python
covid.head(10)
```

```python
covid['Active_cases']=(covid['Confirmed']-(covid['Cured']+covid['Deaths'])) 
covid.head(10)
```

```python
#distribution statewise
statewise=pd.pivot_table(covid,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory',aggfunc=max)
statewise
```

```python
statewise['recovery_rate']=(statewise['Cured']/statewise['Confirmed'])*100
statewise
```

```python
statewise.style.background_gradient(cmap='Accent')
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
#top 10 active cases
top_10_active_cases=(covid.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False))
top_10_active_cases.head()
```

```python
top_10_active_cases=(covid.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False)).reset_index()
top_10_active_cases.head()
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title('top 10 states with active cases')
```

```python
fig=plt.figure(figsize=(16,9))
ax=sns.barplot(
data=top_10_active_cases.iloc[:10],
y='Active_cases',x='State/UnionTerritory',
linewidth=1,edgecolor='red',
    palette='plasma')
plt.xticks(rotation=90)
plt.xlabel('states')
plt.ylabel('total active cases')
plt.show()
```
